```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
# import our 2022 and 2023 fastball records

fvaldez_fb_2022 = pd.read_csv('./data/fvaldez_fb_2022.csv')
fvaldez_fb_2023 = pd.read_csv('./data/fvaldez_fb_2023.csv')
```

```python
# check head of 2022 st

fvaldez_fb_2022.head
```

```python
# plot spin rate vs pitch velo

ax = fvaldez_fb_2022.plot.scatter(x='release_speed', y='release_spin_rate', c='r')
fvaldez_fb_2023.plot.scatter(ax=ax, x='release_speed', y='release_spin_rate')
ax.legend(["2022" , "2023"])
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Spin Rate")
```

```python
# plot horizontal movement vs pitch velocity

ax = fvaldez_fb_2022.plot.scatter(x='release_speed', y='pfx_x', c='r')
fvaldez_fb_2023.plot.scatter(ax=ax, x='release_speed', y='pfx_x')
ax.legend(["2022" , "2023"])
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Horizontal Movement")
```

```python
# plot vertical movement vs pitch velocity

ax = fvaldez_fb_2022.plot.scatter(x='release_speed', y='pfx_z', c='r')
fvaldez_fb_2023.plot.scatter(ax=ax, x='release_speed', y='pfx_z')
ax.legend(["2022" , "2023"])
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Vertical Movement")
```

```python
# look at only cutters

cutter_2023 = fvaldez_fb_2023[fvaldez_fb_2023['pitch_name']=='Cutter']

ax = cutter_2023.plot.scatter(x='release_speed', y='pfx_x')
ax.title.set_text('2023 Cutter Horizontal Movement')
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Horizontal Movement")

ax1 = cutter_2023.plot.scatter(x='release_speed', y='pfx_z')
ax1.title.set_text('2023 Cutter Vertical Movement')
ax1.set_xlabel("Pitch Velocity (mph)")
ax1.set_ylabel("Vertical Movement")
```

```python
# look at only sinkers

sinker_2023 = fvaldez_fb_2023[fvaldez_fb_2023['pitch_name']=='Sinker']

ax = sinker_2023.plot.scatter(x='release_speed', y='pfx_x')
ax.title.set_text('2023 Sinker Horizontal Movement')
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Horizontal Movement")

ax1 = sinker_2023.plot.scatter(x='release_speed', y='pfx_z')
ax1.title.set_text('2023 Sinker Vertical Movement')
ax1.set_xlabel("Pitch Velocity (mph)")
ax1.set_ylabel("Vertical Movement")
```

```python
# look at only four-seam fastballs

ff_2022 = fvaldez_fb_2022[fvaldez_fb_2022['pitch_name']=='4-Seam Fastball']

ax = ff_2022.plot.scatter(x='release_speed', y='pfx_x')
ax.title.set_text('2022 4-Seam Horizontal Movement')
ax.set_xlabel("Pitch Velocity (mph)")
ax.set_ylabel("Horizontal Movement")

ax1 = ff_2022.plot.scatter(x='release_speed', y='pfx_z')
ax1.title.set_text('2022 4-Seam Vertical Movement')
ax1.set_xlabel("Pitch Velocity (mph)")
ax1.set_ylabel("Vertical Movement")
```
